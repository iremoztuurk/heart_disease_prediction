```python
def switch(rows,columns,data):
    global data_remake
    data_remake = [[None] * columns for _ in range(rows)]
    
    for i in range(rows):
        for j in range(columns):
            if j==0:
                data_remake[i][j]=data[i][j]
            elif j==1:
                data_remake[i][j]=data[i][j]
            elif j==2:
                data_remake[i][j]=data[i][j]
            elif j==3:
                data_remake[i][j] = "Male" if data[i][j] ==1 else "Female"
            elif j ==4:
                data_remake[i][j] = "Typical Angina" if data[i][j] ==0 else "Atypical Angina" if data[i][j]==1 else "Non-anginal pain" if data[i][j]==2 else"Asymptomatic"
            elif j==5:
                data_remake[i][j]=data[i][j]
            elif j==6:
                data_remake[i][j]=data[i][j]
            elif j==7:
                data_remake[i][j] = "Yes" if data[i][j] ==1 else "No"
            elif j==8:
                data_remake[i][j] = "Normal" if data[i][j]==0 else "ST-T wave abnormality" if data[i][j]==1 else "Definite left ventricular hypertrophy"
            elif j==9:
                data_remake[i][j]=data[i][j]
            elif j==10:
                data_remake[i][j] = "Yes" if data[i][j]==1 else "No"
            elif j==11:
                data_remake[i][j]=data[i][j]
            elif j==12:
                data_remake[i][j]=data[i][j]
            elif j==13:
                data_remake[i][j]=data[i][j]
            elif j==14:
                data_remake[i][j] = "Normal" if data[i][j]==1 else "Fixed defect" if data[i][j]==2 else "Reversable defect" if data[i][j]==3 else "None"
            elif j==15:
                data_remake[i][j] = "Risk" if data[i][j]==1 else "No Risk"
            elif j==16:
                data_remake[i][j]=data[i][j]

    return data_remake
    
```

```python
def evaluation(main_window,Values,data1,data2,button,button2):
    global target
    config_obj = configparser.ConfigParser()
    config_obj.read("C:/Users/onlinekullanici/Desktop/jupyters/modelconfig.ini"),
    model_param = config_obj["DLmodel"]
    model_path = model_param["model"]
    model = load_model(model_path)
    weight_path = model_param["weights"]
    weights = model.load_weights(weight_path)
    scaler_path = model_param["scaler"]
    connection = connect()
    """model.compile(
        loss = keras.losses.BinaryCrossentropy(from_logits=False),
        optimizer = keras.optimizers.Adam(learning_rate=0.0005),
        metrics = ['accuracy'])"""

    scaler = joblib.load(scaler_path)
    column_names2 = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope",
                 "ca","thal"]
    Values = np.array(Values)
    Values = Values.reshape(1, -1)
    Values = pd.DataFrame(Values,columns=column_names2)
    Values = scaler.transform(Values)
    val_tensor = tf.convert_to_tensor(Values)
    target_arr = model.predict(val_tensor)
    target = target_arr[0][0]
    target_raw = target*100
    threshold = np.float32(model_param["threshold"])
    
    if target > threshold:
        target = 1
        label = Label(main_window, text="The patient is in a risk of heart disease %{:.2f}! (threshold = {})".format(target_raw, threshold), font='Helvetica 10 bold italic', bg= '#BE7B72')
        label.place(x=50, y=830)
    else:
        target = 0
        label = Label(main_window, text="The patient is in a risk of heart disease %{:.2f}! (threshold = {})".format(target_raw, threshold), font='Helvetica 10 bold italic', bg= '#ACE2E1')
        label.place(x=50, y=830)

    data2.insert(11,target)
    cursor = connection.cursor()
    
    if c==1:
        sql2 = "INSERT INTO hasta_bilgi(hasta_id, age, sex,rumus) VALUES(%s, %s, %s,%s)"
        cursor.execute(sql2,data1)
    
    sql3 = "INSERT INTO kalp(cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,target,hasta_id,date_stamp) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
    cursor.execute(sql3,data2)
    
    connection.commit()
    connection.close()
        
    button.place_forget()
    button2.place_forget()
    button3 = button = Button(main_window, text="New Patient", command=close_win)
    button3.place(x=150, y=850)
    button4 = button = Button(main_window, text="Continue", command=same_patient)
    button4.place(x=50, y=850)
    
    
```
