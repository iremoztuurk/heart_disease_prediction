```python
def evaluation(main_window,Values,data1,data2):
    config_obj = configparser.ConfigParser()
    config_obj.read("C:/Users/onlinekullanici/Desktop/jupyters/modelconfig.ini"),
    model_param = config_obj["DLmodel"]
    connection = connect()
    model = load_model(model_param["model"])
    loaded_model = model.load_weights(model_param["weights"])
    """model.compile(
        loss = keras.losses.BinaryCrossentropy(from_logits=False),
        optimizer = keras.optimizers.Adam(learning_rate=0.0005),
        metrics = ['accuracy'])"""

    scaler = joblib.load(model_param["scaler"])
    column_names2 = ["age","sex","cp","trestbps","chol","fbs","restecg","thalach","exang","oldpeak","slope",
                 "ca","thal"]
    Values = np.array(Values)
    Values = Values.reshape(1, -1)
    Values = pd.DataFrame(Values,columns=column_names2)
    Values = scaler.transform(Values)
    val_tensor = tf.convert_to_tensor(Values)
    target_arr = model.predict(val_tensor)
    target = target_arr[0][0]
    threshold = np.float32(model_param["threshold"])
    if target > threshold:
        target = 1
        label = Label(main_window, text=f"The patient is in a risk of heart disease! (threshold = {threshold})", font= 'Helvetica 10 bold italic')
        label.place(x=50,y=830)
    else:
        target = 0
        label = Label(main_window, text=f"The patient is healthy! (threshold = {threshold})", font= 'Helvetica 10 bold italic')
        label.place(x=50,y=830)
    data2.insert(11,target)
    cursor = connection.cursor()
    
    if c==1:
        sql2 = "INSERT INTO hasta_bilgi(hasta_id, age, sex) VALUES(%s, %s, %s)"
        cursor.execute(sql2,data1)
    
    sql3 = "INSERT INTO kalp(cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal,target,hasta_id) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
    cursor.execute(sql3,data2)
    
    connection.commit()
    connection.close()
     
   
    
```
